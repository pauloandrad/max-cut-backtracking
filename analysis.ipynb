## O(2^n) Exponencial

```python
import subprocess
import time
import matplotlib.pyplot as plt

nodes = [2, 4, 6, 8, 10, 12, 14, 16]
times = []

for n in nodes:
    print(f"Rodando script para n={n}...")

    start = time.perf_counter()
    process = subprocess.run(['python', 'brute_force.py', str(n)],capture_output=True, text=True)
    end = time.perf_counter()
    times.append(end - start)

plt.plot(nodes, times, 'g-s')
plt.title("Performance do Arquivo Completo (Max Cut)")
plt.xlabel("Valor de n (max_graph)")
plt.ylabel("Tempo de Execução (s)")
plt.show()
```

### Brute force: complexidade exponencial 2n testa cada possíbilidade o que representa o dobro de trabalho a cada passo

### O(n^2) Quadrática


```python
import subprocess
import time
import matplotlib.pyplot as plt

nodes = [10, 20, 30, 40, 60, 70, 80, 90, 100, 110, 120]
times = []

for n in nodes:
    print(f"Rodando script para n={n}...")

    start = time.perf_counter()
    process = subprocess.run(['python', 'backtracking.py', str(n)],capture_output=True, text=True)
    end = time.perf_counter()
    times.append(end - start)

plt.plot(nodes, times, 'g-s')
plt.title("Performance do Arquivo Completo (Max Cut)")
plt.xlabel("Valor de n (max_graph)")
plt.ylabel("Tempo de Execução (s)")
plt.show()
```

### Backtracking poda simples: Quadratico n^2 (Com resalva)

### se tratando de um grafo circular, o backtracking conseguiu lidar bem, entretanto, em cenários mais realistas grafos tendem a ser "Aleatórios" o que torna podas mais complicadas ou até impossíveis.
